```python
import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams["font.family"]="SimHei" # 中文
path=r"C:\Users\gaoxiang\Documents\GitHub\django\数据集\test1.csv"
pwd = os.getcwd()
os.chdir(os.path.dirname(path))
unrate = pd.read_csv(os.path.basename(path),encoding="gbk")
#os.chdir(pwd)
unrate.info()
```

# 分类数据

## 散点图 

### stripplot

```python
sns.stripplot(x="乘客等级",y="票价",data=unrate,hue="性别",jitter=True)
```

### swarmplot 

```python
sns.swarmplot (x="乘客等级",y="票价",data=unrate,hue="性别")
```

## 盒形图

### boxplot

```python
sns.boxplot  (x="乘客等级",y="票价",data=unrate,hue="性别", dodge = False)#取消合并
```

### violinplot

```python
sns.violinplot   (x="乘客等级",y="票价",data=unrate,hue="性别", dodge = False)#取消合并
```

## 条形图

### barplot#期望

```python
sns.barplot    (x="乘客等级",y="票价",data=unrate,hue="性别")
```

### countplot  计数

```python
sns.countplot     (x="乘客等级",data=unrate,hue="性别")
```

## 点线图

### pointplot

```python
sns.pointplot  (x="乘客等级",y="票价",data=unrate,hue="性别")
```

# 多面板

## factorplot 

```python
sns.factorplot(x="乘客等级",y="票价",data=unrate,col ="性别", kind = "swarm" )
```

## PairGrid

```python
g=sns.PairGrid(unrate,x_vars=["乘客等级", "性别"],
                 y_vars=["票价", "年龄"],
          )
g.map(sns.swarmplot , palette="pastel");
```
