## stats

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

### 常用方法
* rvs: 随机变量
* pdf: 概率密度函数
* cdf: 累积分布函数
* sf:  生存函数(1-CDF)
* ppf: 百分比点函数(CDF)的逆
* isf: 逆生存函数(科幻的逆)
* stats: 返回的意思是,方差,(费舍尔)倾斜,或峰度(费雪的)
* moment:偏心的时刻的分布

##### 概率密度函数

```python
n=np.arange(-1,1,0.01)
y=stats.norm.pdf(n,0,1)
plt.plot(n,y,"o-")
plt.show()
```

```python
stats.norm.rvs(size=100)
stats.norm.cdf(0)
stats.norm.ppf(0.5)
mean, var, skew, kurt =stats. norm.stats(moments='mvsk')
stats.norm.mean(), stats.norm.std(),stats.norm.var()
stats.describe(stats.norm(0,1))
```

```python
stats.norm.stats(loc=3, scale=4, moments="mv")
```

* plt.plot(x,y,fmt,…) 绘制一个坐标图
* plt.boxplot(data,notch,position) 绘制一个箱形图
* plt.bar(left,height,width,bottom) 绘制一个条形图

```python
import numpy
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
import numpy as np 
from scipy.stats import norm
import matplotlib.pyplot as plt
y=np.linspace(-5,5,5000)
x=norm.pdf(y)
plt.plot(y,x,'1')
plt.vlines([0,0.2,0.5],x.min(),x.max(),linestyles="dashed")
plt.fill_between(np.linspace(-2,2),norm.pdf(np.linspace(-2,2)), facecolor = "yellow")
plt.show()
```
